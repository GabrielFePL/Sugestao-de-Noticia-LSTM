# Rede Neural para Predição de Notícias Sugeridas

## Base Histórica de Dados Simulada

### Importação das Bibliotecas

```python
import numpy as np
import pandas as pd
import random
from datetime import datetime, timedelta
```

### Definição de Subcategorias das Notícias

```python
categories = [
    'Vagas de emprego', 'Carreira e Desenvolvimento Profissional', 'Dicas de Entrevistas e Currículos',
    'Cursos Profissionalizantes', 'Certificações Técnicas', 'Ferramentas e Habilidades Práticas',
    'Futebol', 'Esportes Aquáticos', 'Esportes Radicais',
    'Filmes e Séries', 'Música e Shows', 'Cultura Pop e Celebridades',
    'Mercado Financeiro', 'Políticas Econômicas', 'Investimentos e Finanças Pessoais'
]
```

### Definição do Número de Usuários da Simulação

```python
num_users = 2000
```

### Definição da Probabilidade de Acesso às Subcategorias

```python
probabilities = [
    0.12,  # Vagas de emprego
    0.10,  # Carreira e desenvolvimento profissional
    0.08,  # Dicas de entrevistas e currículos
    0.10,  # Cursos profissionalizantes
    0.09,  # Certificações técnicas
    0.08,  # Ferramentas e habilidades práticas
    0.05,  # Futebol
    0.04,  # Esportes aquáticos
    0.03,  # Esportes radicais
    0.06,  # Filmes e séries
    0.05,  # Música e shows
    0.04,  # Cultura pop e celebridades
    0.06,  # Mercado financeiro
    0.05,  # Políticas econômicas
    0.05   # Investimentos e finanças pessoais
]
```

### Definição de Data de Acesso Aleatória para os Registros

```python
def random_date(start_date, end_date):
  return start_date + timedelta(days = random.randint(0, int((end_date - start_date).days)))

start_date = datetime(2023, 8, 23)
end_date = datetime(2024, 8, 23)
```

### Geração da Base de Dados Histórica Simulada

```python
data = []

for user_id in range(1, num_users + 1):
  # Contagem de acessos aleatória para as subcategorias
  total_accesses = np.random.randint(25, 75)

  # Distribuição das probabilidades de acesso
  user_accesses = np.random.multinomial(total_accesses, probabilities)

  # Geração de data aleatória para os registros
  access_date = random_date(start_date, end_date)

  # Geração de Registro com AccessDate e SubcategoriesCounts
  row = [access_date] + user_accesses.tolist()
  data.append(row)

# Definição das Colunas
columns = ['AccessDate'] + categories

# Conversão para DataFrame
df = pd.DataFrame(data, columns = columns)
```
